## 2.1 k-means: a first version

```python
# Self-defined k-means algorithm

def mykmeans(data,max_k,max_iter=50):
    length = data.shape[1]
    num = data.shape[0]
    rss_record = []
    k_clusters = [[] for i in xrange(max_k)]
    centroids_list = [[] for i in xrange(max_k)]
    random.seed(11031991)
    k_idx = random.sample(xrange(num),max_k)
    for i in xrange(max_k):
        k_clusters[i].append(data[k_idx[i]])
        centroids_list[i].append(data[k_idx[i]])
    first_loop = True
    
    for l in xrange(max_iter):
        rss_list = []
        for m in range(num):
            distance = [np.sqrt(sum((data[m]-np.array(k_clusters[i][0]))**2)) for i in xrange(max_k)]
            idx = distance.index(min(distance))
            if first_loop:
                if m not in k_idx:
                    k_clusters[idx].append(data[m])
            else:
                k_clusters[idx].append(data[m])
        if l != (max_iter-1):
            if first_loop:
                mean = [np.mean(k_clusters[i],axis=0) for i in xrange(max_k)]
                for n in xrange(max_k):
                    rss_list.append(np.sum([sum((x-mean[n])**2) for x in k_clusters[n]]))
                rss = np.sum(rss_list)
                rss_record.append(rss)
                k_clusters = [[] for i in xrange(max_k)]
                for i in xrange(max_k):
                    k_clusters[i].append(mean[i])
                    centroids_list[i].append(mean[i])
            else:
                mean = [np.mean(k_clusters[i][1:],axis=0) for i in xrange(max_k)]
                for n in xrange(max_k):
                    rss_list.append(np.sum([sum((x-mean[n])**2) for x in k_clusters[n]]))
                rss = np.sum(rss_list)
                rss_record.append(rss)
                k_clusters = [[] for i in xrange(max_k)]
                for i in xrange(max_k):
                    k_clusters[i].append(mean[i])
                    centroids_list[i].append(mean[i])
        else:
            mean = [np.mean(k_clusters[i],axis=0) for i in xrange(max_k)]
            for i in xrange(max_k):
                    centroids_list[i].append(mean[i])
                    
        first_loop = False

    for i in xrange(max_k):
        k_clusters[i] = k_clusters[i][1:]

    rss_list = []
    for n in xrange(max_k):
        rss_list.append(np.sum([np.sqrt(sum((x-mean[n])**2)) for x in k_clusters[n]]))
    rss = np.sum(rss_list)
    rss_record.append(rss)
    
    return rss,k_clusters,centroids_list,rss_record
```

## 2.2   k-means: a better version

```python
# Implementation of a better k-means

def mykmeans_multi(data,max_k,max_iter=50,num_r=100):
    r_record = []
    centroids_record = []
    rss_record = []
    clusters_record = []
    for r in xrange(num_r):
        rss,k_clusters,centroids_list,rss_list = mykmeans(X_train,max_k,max_iter=50)
        r_record.append(rss)
        centroids_record.append(centroids_list)
        clusters_record.append(k_clusters)
        rss_record.append(rss_list)
    idx = r_record.index(min(r_record)) 
    print 'The best run is : %d' %(idx+1)
    
    return min(r_record),clusters_record[idx],centroids_record[idx],rss_record[idx] 
```

## 2.3 k-means++

    Choose one new data point at random as a new center, using a weighted probability distribution where a point x is chosen with probability proportional to $D(x)^2$.

```python
# Implementation of k-means++

def mykmeans_pp(data,max_k,max_iter=50):
    length = data.shape[1]
    num = data.shape[0]
    mean = [[] for i in xrange(max_k)]
    rss_record = []
    k_clusters = [[] for i in xrange(max_k)]
    centroids_list = [[] for i in xrange(max_k)]
    random.seed(11031991)
    k_idx = random.sample(xrange(num),max_k)
    for i in xrange(max_k):
        k_clusters[i].append(data[k_idx[i]])
        centroids_list[i].append(data[k_idx[i]])

    for l in xrange(max_iter):
        rss_list = []
        distance_record = []
        for m in range(num):
            distance = [np.sqrt(sum((data[m]-np.array(k_clusters[i][0]))**2)) for i in xrange(max_k)]
            distance_record.append(distance)
            print distance
            idx = distance.index(min(distance))
            if m not in k_idx:
                k_clusters[idx].append(data[m])
        if l != (max_iter-1):
            prob = np.array(distance_record)**2/sum(np.array(distance_record)**2)
            for i in xrange(max_k):
                k_idx[i] = np.random.choice(xrange(data.shape[0]), p=prob[:,i])
                mean[i] = data[k_idx]
            for n in xrange(max_k):
                rss_list.append(np.sum([np.sqrt(sum((x-mean[n])**2)) for x in k_clusters[n]]))
            rss = np.sum(rss_list)
            rss_record.append(rss)
            k_clusters = [[] for i in xrange(max_k)]
            for i in xrange(max_k):
                k_clusters[i].append(mean[i])
                centroids_list[i].append(mean[i])
        else:
            prob = np.array(distance_record)**2/sum(np.array(distance_record)**2)
            for i in xrange(max_k):
                    k_idx[i] = np.random.choice(xrange(data.shape[0]), p=prob[:,i])
                    mean[i] = data[k_idx]
            for i in xrange(max_k):
                    centroids_list[i].append(mean[i])
        print 'sd'
  
    rss_list = []
    for n in xrange(max_k):
        rss_list.append(np.sum([np.sqrt(sum((x-mean[n])**2)) for x in k_clusters[n]]))
    rss = np.sum(rss_list)
    rss_record.append(rss)
    
    return rss,k_clusters,centroids_list,rss_record
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn import metrics
from sklearn.grid_search import GridSearchCV
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn import metrics
from sklearn.datasets import fetch_20newsgroups
```

```python
categories = ['alt.atheism', 'soc.religion.christian','comp.graphics', 'sci.med']
twenty_train = fetch_20newsgroups(subset='train',categories=categories, shuffle=True, random_state=42)
```

```python
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(twenty_train.data)
X_train_counts.shape
```

```python
count_vect.vocabulary_.get(u'algorithm')
```

```python
tf_transformer = TfidfTransformer(use_idf=False).fit(X_train_counts)
X_train_tf = tf_transformer.transform(X_train_counts)
X_train_tf.shape
```

```python
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
X_train_tfidf.shape
```

```python
clf = KNeighborsClassifier(n_neighbors=10).fit(X_train_tfidf, twenty_train.target)
```

```python
docs_new = ['God is love', 'OpenGL on the GPU is fast']
X_new_counts = count_vect.transform(docs_new)
X_new_tfidf = tfidf_transformer.transform(X_new_counts)

predicted = clf.predict(X_new_tfidf)

for doc, category in zip(docs_new, predicted):
    print('%r => %s' % (doc, twenty_train.target_names[category]))
```

## Build a pipeline

```python
text_clf = Pipeline([('vect', CountVectorizer()),('tfidf', TfidfTransformer()),('clf', KNeighborsClassifier())])
text_clf = text_clf.fit(twenty_train.data, twenty_train.target)
twenty_test = fetch_20newsgroups(subset='test',categories=categories, shuffle=True, random_state=42)
docs_test = twenty_test.data
predicted = text_clf.predict(docs_test)
np.mean(predicted == twenty_test.target)            
```

```python
print(metrics.classification_report(twenty_test.target, predicted,target_names=twenty_test.target_names))
```

```python
parameters = {'vect__ngram_range': [(1, 1), (1, 2)],'tfidf__use_idf': (True, False),'clf__n_neighbors': (2,4,6,8,10)}
gs_clf = GridSearchCV(text_clf, parameters, n_jobs=-1)
gs_clf = gs_clf.fit(twenty_train.data[:400], twenty_train.target[:400])
```

```python
"""Build a sentiment analysis / polarity model

Sentiment analysis can be casted as a binary text classification problem,
that is fitting a linear classifier on features extracted from the text
of the user messages so as to guess wether the opinion of the author is
positive or negative.

In this examples we will use a movie review dataset.

"""
# Author: Olivier Grisel <olivier.grisel@ensta.org>
# License: Simplified BSD


from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.grid_search import GridSearchCV
from sklearn.datasets import load_files
from sklearn.cross_validation import train_test_split
from sklearn import metrics


if __name__ == "__main__":
    # NOTE: we put the following in a 'if __name__ == "__main__"' protected
    # block to be able to use a multi-core grid search that also works under
    # Windows, see: http://docs.python.org/library/multiprocessing.html#windows
    # The multiprocessing module is used as the backend of joblib.Parallel
    # that is used when n_jobs != 1 in GridSearchCV

    # the training data folder must be passed as first argument
    dataset = load_files('txt_sentoken', shuffle=False)
    print("n_samples: %d" % len(dataset.data))

    # split the dataset in training and test set:
    docs_train, docs_test, y_train, y_test = train_test_split(
        dataset.data, dataset.target, test_size=0.25, random_state=None)

    # TASK: Build a vectorizer / classifier pipeline that filters out tokens
    # that are too rare or too frequent
    text_clf = Pipeline([('vect',TfidfVectorizer(stop_words='english')),('tfidf',TfidfTransformer()),('clf',KNeighborsClassifier())])

    # TASK: Build a grid search to find out whether unigrams or bigrams are
    # more useful.
    parameters = {'vect__ngram_range': [(1, 1), (1, 2)],
                  'tfidf__use_idf': (True, False),'clf__n_neighbors': (2,4,6,8,10)}
    movie_clf = GridSearchCV(text_clf,parameters,n_jobs=-1)
    
    # Fit the pipeline on the training set using grid search for the parameters
    movie_clf = movie_clf.fit(docs_train,y_train)

    # TASK: print the cross-validated scores for the each parameters set
    # explored by the grid search
    best_parameters, score, _ = max(movie_clf.grid_scores_, key=lambda x: x[1])
    for param_name in sorted(parameters.keys()):
        print("%s: %r" % (param_name, best_parameters[param_name]))

    # TASK: Predict the outcome on the testing set and store it in a variable
    # named y_predicted
    y_predicted = movie_clf.predict(docs_test)

    # Print the classification report
    print(metrics.classification_report(y_test, y_predicted,
                                        target_names=dataset.target_names))

    # Print and plot the confusion matrix
    cm = metrics.confusion_matrix(y_test, y_predicted)
    print(cm)


    plt.matshow(cm)
    plt.show()
```
